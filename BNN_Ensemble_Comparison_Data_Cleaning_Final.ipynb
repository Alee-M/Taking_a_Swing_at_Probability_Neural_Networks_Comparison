```
#Import necessary packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

###2017 Data


Start: 4/2/2017
ASG: 7/11/2017
EOS: 10/1/2017

```
#ASG data represents SO/PA at the midpoint of the baseball season (the All-Star Break)
ASG2017 = pd.read_csv(dir + '2017_Start_ASG_160_AB.csv') #Load in CSV (direct export from FanGraphs)
ASG2017 = ASG2017[['playerId','SO','PA']] #Remove unnecessary columns
ASG2017
```

```
#EOS data represents SO/PA with full season data
EOS2017 = pd.read_csv(dir + '2017_Start_EOS_160_AB.csv') #Load in CSV (direct export from FanGraphs)
EOS2017 = EOS2017[['playerId','SO','PA']] #Remove unnecesaary columns
EOS2017['SO/PA'] = EOS2017['SO']/EOS2017['PA'] #Calculate strikeout rate
EOS2017=EOS2017.drop(columns=['PA','SO']) #Drop columns used for calculation
EOS2017
```

```
#Create merged data for 2017
data2017 = pd.merge(ASG2017, EOS2017, on=['playerId'], how='inner') #Merge data at All-Star_Break with data at End of Season, inner join -> only players in both datasets
data2017=data2017.dropna() #Drop rows with blank values
data2017
```

### Merging All Data

```
#Make dataframe with all years
Full_data=pd.concat([data2017, data2018, data2019, data2021, data2022]) #Concatenate all previous dataframes
Full_data.rename(columns = {'SO/PA':'EOS_SO/PA'}, inplace = True) #Rename to differentiate stats
Full_data
```

### Define Inputs and Outputs, Test/Train Split

```
#Define X and Y variables
x=Full_data[['playerId','SO','PA']] #Only variables for ASG SO/PA
y=Full_data[['playerId','EOS_SO/PA']] #End of Season SO/PA
```

```
#Test/Train Split
X_train, X_test, Y_train, Y_test = train_test_split(x, y)
```
